```python
#In the below script, I will solve the harmonic oscillator numerically using a simple implementation of the leap frog method.


import numpy as np
import matplotlib.pyplot as plt
```

```python
#===System Parameters===
k = 1.0
x0 = 1.0
v0 = 0.0
dt = 0.1
t_max = 15

#===Time array===
t = np.arange(0, t_max, dt) #All possible time steps are created here

#===X & T Array===
x = np.zeros(len(t))
v_half = np.zeros(len(t))

#===Initial conditions===
x[0] = x0
v_half[0] = v0 + 0.5 * dt * (-k * x0) #-kx term from the harmonic ode

#===Leapfrog integration===
for i in range(1, len(t)): #integration bounds
    x[i] = x[i-1] + dt * v_half[i-1]
    v_half[i] = v_half[i-1] + dt * (-k * x[i])

#full velocity update
v = np.zeros(len(t))
v[0] = v0
for i in range(1, len(t)):
    v[i] = v_half[i-1] + 0.5 * dt * (-k * x[i])

# Plotting the results
plt.plot(t, x, label='Position')
plt.plot(t, v, label='Velocity')
plt.xlabel('Time')
plt.ylabel('Position / Velocity')
plt.legend()
plt.show()
```
